```python
worldcup_teams = ['England', ' South Africa', '', 'West Indies',
            'Pakistan', 'New Zealand', 'Sri Lanka', 'Afghanistan',
            'Australia', 'Bangladesh', 'India']

# Filtering matches involving only teams in the 'worldcup_teams' list
df_teams_1 = results[results['Team_1'].isin(worldcup_teams)]  # Extracts matches with Team_1 in 'worldcup_teams'
df_teams_2 = results[results['Team_2'].isin(worldcup_teams)]  # Extracts matches with Team_2 in 'worldcup_teams'

# Combining the filtered dataframes
df_teams = pd.concat((df_teams_1, df_teams_2))  # Concatenates the two dataframes vertically

df_teams.drop_duplicates()  # Removes duplicate rows if any

df_teams.count()  # Counts the number of rows in the final dataframe
df_teams_2019 = df_teams.drop(['Date','Margin', 'Ground'], axis=1)

df_teams_2019 = df_teams_2019.reset_index(drop=True)

# Set a value of 1 in the 'winning_team' column for rows where the 'Winner' column is the same as the 'Team_1' column.
df_teams_2019.loc[df_teams_2019.Winner == df_teams_2019.Team_1,'winning_team']=1

# Set a value of 2 in the 'winning_team' column for rows where the 'Winner' column is the same as the 'Team_2' column.
df_teams_2019.loc[df_teams_2019.Winner == df_teams_2019.Team_2, 'winning_team']=2

# Drop the 'winning_team' column from the DataFrame.
df_teams_2019 = df_teams_2019.drop(['winning_team'], axis=1)

# Display the first few rows of the modified DataFrame using df_teams_2019.head().
df_teams_2019.head()
from sklearn.model_selection import train_test_split
final = pd.get_dummies(df_teams_2019, prefix=['Team_1', 'Team_2'], columns=['Team_1', 'Team_2'])

X = final.drop(['Winner'], axis=1)  # Extracting features by dropping the 'Winner' column
y = final["Winner"]  # Assigning the 'Winner' column as the target variable

# Split the data into training and testing sets using a test size of 20% and a random state of 42.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Import the Random forest module from sklearn
from sklearn.ensemble import RandomForestClassifier

# Create an Random forest classifier with a linear kernel
rf = RandomForestClassifier(n_estimators=100, max_depth=20,
                              random_state=0)

# Fit the classifier on the training data
rf.fit(X_train, y_train)

# Make predictions on the test data
pred = rf.predict(X_test)

# Calculating training set score:
score = rf.score(X_train, y_train)

# Calculating test set score:
score2 = rf.score(X_test, y_test)

# Printing the training set accuracy with 3 decimal places
print("Training set accuracy: ", '%.3f'%(score))

# Printing the test set accuracy with 3 decimal places
print("Test set accuracy: ", '%.3f'%(score2))
```

```python
# List of teams qualified for the semi-finals based on their wins
semi_finals = [('Pakistan', 'Australia'),
               ('England', 'India')]

def clean_and_predict(matches, ranking, final, rf):
    # Initialize an empty list to store the positions of teams in the ranking
    positions = []

    # Loop through the list of matches and add the rankings of the teams to the positions list
    for match in matches:
        positions.append(ranking.loc[ranking['Team_name'] == match[0], 'Team_ranking'].iloc[0])
        positions.append(ranking.loc[ranking['Team_name'] == match[0], 'Team_ranking'].iloc[0])

    # Initialize an empty list to store the match predictions
    predictions_set = []

    # Initialize indices i and j to 0 for iteration
    i = 0
    j = 0

    # Loop through the positions list to create match predictions
    while i < len(positions):
        prediction_dict = {}
        # Compare the positions of the two teams in the match
        # and add the corresponding team names to the prediction dictionary
        if positions[i] < positions[i + 1]:
            prediction_dict.update({'Team_1': matches[j][0], 'Team_2': matches[j][1]})
        else:
            prediction_dict.update({'Team_1': matches[j][1], 'Team_2': matches[j][0]})

        # Add the prediction dictionary to the predictions_set list
        predictions_set.append(prediction_dict)
        i += 2
        j += 1

    # Convert the predictions_set list into a pandas DataFrame
    predictions_set = pd.DataFrame(predictions_set)

    # Make a backup of the predictions_set DataFrame
    backup_predictions_set = predictions_set

    # Convert categorical variables in the predictions_set DataFrame into dummy variables
    predictions_set = pd.get_dummies(predictions_set, prefix=['Team_1', 'Team_2'], columns=['Team_1', 'Team_2'])

    # Identify missing columns in predictions_set compared to the final DataFrame
    missing_cols = set(final.columns) - set(predictions_set.columns)

    # Add the missing columns to the predictions_set DataFrame and set their values to 0
    for col in missing_cols:
        predictions_set[col] = 0

    # Reorder predictions_set columns to match the final DataFrame's order
    predictions_set = predictions_set[final.columns]

    # Drop the 'Winner' column from the predictions_set DataFrame
    predictions_set = predictions_set.drop(['Winner'], axis=1)

    # Use the provided classifier (rf) to make predictions on the predictions_set DataFrame
    predictions = rf.predict(predictions_set)

    # Loop through the predictions_set DataFrame and print the match predictions and winners
    for i in range(len(predictions_set)):
        print(backup_predictions_set.iloc[i, 1] + " vs " + backup_predictions_set.iloc[i, 0])
        if predictions[i] == 1:
            print("Winner: " + backup_predictions_set.iloc[i, 1])
        else:
            print("Winner: " + backup_predictions_set.iloc[i, 0])
        print("")

# Call the clean_and_predict function with the provided arguments
clean_and_predict(semi_finals, ranking, final, rf)
```

```python
import pandas as pd

# Read the bat and ball datasets
df_bat = pd.read_csv('/content/drive/MyDrive/data_mining/bat.csv')
df_ball = pd.read_csv('/content/drive/MyDrive/data_mining/ball.csv')

# Merge the datasets on "Player" and "Country" columns
merged_df = pd.merge(df_bat[['Player', 'Country', 'Ave']],
                     df_ball[['Player', 'Country', 'Ave']],
                     on=['Player', 'Country'],
                     how='outer',
                     suffixes=('_bat', '_ball'))

# Replace NaN values with 0
merged_df.fillna(0, inplace=True)



import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense



# Encode categorical features ('Country' column) using LabelEncoder
label_encoder = LabelEncoder()
merged_df['Country'] = label_encoder.fit_transform(merged_df['Country'])

# Filter players from the top 2 teams (Australia and India)
finals_teams = ['Australia', 'India']

# Create a dictionary to store the predicted playing 11 for each team
predicted_playing_11 = {}

# Iterate over each team
for team in finals_teams:
    # Combine the data for the selected team
    selected_team_df = merged_df[merged_df['Country'] == label_encoder.transform([team])[0]]

    # Define features and target variable
    X = selected_team_df[['Ave_bat', 'Ave_ball']]
    y = pd.Series(1, index=X.index)

    # Standardize the features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Build the neural network model
    model = Sequential()
    model.add(Dense(32, input_dim=2, activation='relu'))
    model.add(Dense(16, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model
    model.fit(X_scaled, y, epochs=50, batch_size=32, verbose=0)

    # Predict probabilities for each player
    predicted_probabilities = model.predict(X_scaled)

    # Select the top 11 players with the highest probabilities
    selected_players_indices = np.argsort(predicted_probabilities[:, 0])[-11:][::-1]
    selected_players = selected_team_df.iloc[selected_players_indices]

    # Ensure there are at least 3-4 bowlers and 1-2 all-rounders in the team
    bowlers = selected_players[selected_players['Ave_ball'] > 0][:4]
    all_rounders = selected_players[(selected_players['Ave_bat'] > 0) & (selected_players['Ave_ball'] > 0)][:2]
    remaining_players = selected_players[selected_players['Ave_bat'] > 0][:5]

    # Combine the selected players
    final_playing_11 = pd.concat([bowlers, all_rounders, remaining_players])

    # Store the predicted playing 11 in the dictionary
    predicted_playing_11[team] = final_playing_11

# Display the predicted playing 11 for each team
for team, players_df in predicted_playing_11.items():
    print(f"Predicted playing 11 for {team}:")
    print(players_df[['Player', 'Country', 'Ave_bat', 'Ave_ball']])
    print("\n")
```
